```python
import random
import time
import matplotlib.pyplot as plt


world_states = ["A", "B"]

number_of_iterations = 25
number_of_timesteps = 50

success_list = []

class Agent:
    def __init__(self):
        self.learner_agent = {"A": ["action1", "action2"], "B": ["action1", "action2"]}

    def select_action(self, current_world_state):
        action = random.choice(self.learner_agent[current_world_state])
        return action
    def receive_reward(self, current_world_state, action):
        self.learner_agent[current_world_state].append(action)
        # Placeholder function for receiving reward
        #print("NEw learner agent for ", agent, self.learner_agent)
        pass

agent1 = Agent()
agent2 = Agent()

action1_total = 0
action2_total = 0


for iteration in range(number_of_iterations):
    success_rate = 0
    number_incorrect = 0
    for t in range(number_of_timesteps):
        current_world_state = random.choice(world_states)
        #print("Current world state:", current_world_state)

        agent1_action = agent1.select_action(current_world_state)
        agent2_action = agent2.select_action(current_world_state)

        if agent1_action == agent2_action:
            #print("Agents successfully matched actions!")
            if agent1_action == "action1":
                action1_total += 1
            if agent1_action == "action2":
                action2_total += 1
            agent1.receive_reward(current_world_state, agent1_action)
            agent2.receive_reward(current_world_state, agent1_action)
            success_rate += 1
            # Both agents receive a reward for successful communication
        else:
            number_incorrect += 1
            #print("Agents failed to match actions!")

        time.sleep(0.01)

    #print("Success rate for iteration:", success_rate / number_of_timesteps)
    success_list.append(success_rate/number_of_timesteps)


print("Number incorrect on last iteration", number_incorrect)
print("Totals of action 1 vs action 2", action1_total, action2_total)
print("Percentage of action 1", action1_total / (action1_total + action2_total))


plt.plot(range(1, number_of_iterations + 1), success_list)
plt.xlabel('Iteration')
plt.ylabel('Success Rate')
plt.title('Success Rate at Each Iteration')
plt.show()
print(success_list)
```

```python
import random
import time
import matplotlib.pyplot as plt

world_states = ["A", "B", "C","D"]

number_of_iterations = 25
number_of_timesteps = 50

success_list = []

class Agent:
    def __init__(self):
        self.learner_agent = {"A": ["action1", "action2", "action3", "action4"], "B": ["action1", "action2", "action3", "action4"], "C": ["action1", "action2", "action3", "action4"], "D": ["action1", "action2", "action3", "action4"]}

    def select_action(self, current_world_state):
        return random.choice(self.learner_agent[current_world_state])
    def receive_reward(self, current_world_state):
        self.learner_agent[current_world_state].append(current_world_state)
        # Placeholder function for receiving reward
        #print("NEw learner agent for ", agent, self.learner_agent)
        pass

agent1 = Agent()
agent2 = Agent()

action1_total = 0
action2_total = 0
action3_total = 0
action4_total = 0
for iteration in range(number_of_iterations):
    success_rate = 0
    for t in range(number_of_timesteps):
        current_world_state = random.choice(world_states)
        #print("Current world state:", current_world_state)

        agent1_action = agent1.select_action(current_world_state)
        agent2_action = agent2.select_action(current_world_state)

        #print("Agent A selects action:", action_A)
        #print("Agent B selects action:", action_B)

        if agent1_action == agent2_action:
            #print("Agents successfully matched actions!")

            # record number of times each action is involved in a success
            if agent1_action == "action1":
                action1_total += 1
            if agent1_action == "action2":
                action2_total += 1
            if agent1_action == "action3":
                action3_total += 1
            if agent1_action == "action4":
                action4_total += 1

            agent1.receive_reward(current_world_state)
            agent2.receive_reward(current_world_state)
            success_rate += 1
            # Both agents receive a reward for successful communication

        else:
            pass
            #print("Agents failed to match actions!")

        time.sleep(0.01)

    print("Success rate for iteration:", success_rate / number_of_timesteps)
    success_list.append(success_rate/number_of_timesteps)

print("Percentage of action 1", action1_total / (action1_total + action2_total + action3_total + action4_total))
print("Percentage of action 2", action2_total / (action1_total + action2_total + action3_total + action4_total))
print("Percentage of action 3", action3_total / (action1_total + action2_total + action3_total + action4_total))
print("Percentage of action 4", action4_total / (action1_total + action2_total + action3_total + action4_total))

plt.plot(range(1, number_of_iterations + 1), success_list)
plt.xlabel('Iteration')
plt.ylabel('Success Rate')
plt.title('Success Rate at Each Iteration')
plt.show()
print(success_list)
```
