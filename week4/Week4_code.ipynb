#### Filtering DataFrames
_filtering_ means keeping some parts and throwing away others.
Typically based on rows values that match some conditions
Pandas `DataFrame` can  be filtered according values of a column by using a boolean expression, for example:

````python
WC_DF['capital'] == 'admin'
````

````python
filtered_DF = WC_DF[ WC_DF['capital'] == 'admin'] # This keeps only administrative capitals
filtered_DF.head()
````

How it works:
*  `DF['label']` (where `DF` is any `DataFrame`), gives a `Series` corresponding
    to    the `'label'` column of `DF.
    
*  `a_series == val` or a Boolean operator (such `==`, `<` etc.) is applied to a `Series` object
   the result is actually a `Series` of Boolean values `True` for each element where
   the original `Series` satisfies `element == val`, and `False` for the rest.
   
* `DF[ bool_series ]`, is an operation, where a boolean
   series is given as a selection argument to the `DataFrame`. It will return 
   a new DF, containing all the rows of `DF` for which `bool_series` has the 
   value `True`

### B

There are several cities that have the same name. Write a function `num_cities_occurring_n_times(n)`, such that:
    
* The argument `n` is an integer;
* The returns an integer, which is the number of different city names that occur `n` times in the `worldcities` dataset. 
* The function should consider the city name to be the value in the `city` column. In other words the name in the form that may contain non-ASCII characters.

````python
WC_DF
````

````python
# B
def num_cities_occurring_n_times(n):
    cities = []
    n_times = {}
    for i, row in WC_DF.iterrows():
        if row['city'] not in n_times:
            n_times[row['city']] = 1
        else:
            n_times[row['city']] += 1
    for city, v in n_times.items():
        if v == n:
            cities.append(city)
    return len(cities)

for i in range(20):
    print(i, num_cities_occurring_n_times(i))
````

````python
def num_cities_occurring_n_times(n):
    count_list = []
    num = 0
    city_list = []
    for city in WC_DF['city']:
        count = 0
        if city not in city_list:
            city_list.append(city)
            for repeat in WC_DF['city']: 
                if city == repeat:
                    count += 1
            count_list.append(count)
    for i in count_list:
        if i == n:
            num += 1
    return num

num_cities_occurring_n_times(2)
````
